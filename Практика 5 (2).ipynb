### 1. Для $ n $ точек примените написанную в предыдущей работе функцию прогноза линейной регрессии. Для полученного прогноза с помощью собственных функций вычислите:
- MSE;
- RMSE;
- MAE;
- Коэффициент детерминации;
- MAPE;
- WAPE;
- BadFreq.

```python
import numpy as np

def get_mse(y_true, y_pred):
    
    return np.mean((y_true - y_pred)**2)

def get_rmse(y_true, y_pred):
   
    return np.sqrt(get_mse(y_true, y_pred))

def get_mae(y_true, y_pred):
  
    return np.mean(np.abs(y_true - y_pred))

def get_r2(y_true, y_pred):
   
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    return 1 - (ss_res / ss_tot)

def get_mape(y_true, y_pred):
  
    return np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

def get_wape(y_true, y_pred):
 
    return np.sum(np.abs(y_true - y_pred)) / np.sum(y_true) * 100

def get_bad_freq(y_true, y_pred, threshold=0.1):

    errors = np.abs((y_true - y_pred) / (y_true + 1e-10))
    bad_predictions = errors > threshold
    return np.mean(bad_predictions) * 100




y_true = np.array([10, 20, 30, 40, 50])
y_pred = np.array([11, 18, 32, 39, 55]) 

print(f"MSE:      {get_mse(y_true, y_pred):.2f}")
print(f"RMSE:     {get_rmse(y_true, y_pred):.2f}")
print(f"MAE:      {get_mae(y_true, y_pred):.2f}")
print(f"R^2:      {get_r2(y_true, y_pred):.4f}")
print(f"MAPE:     {get_mape(y_true, y_pred):.2f}%")
print(f"WAPE:     {get_wape(y_true, y_pred):.2f}%")
print(f"BadFreq:  {get_bad_freq(y_true, y_pred, 0.1):.2f}% (порог 10%)")
```
