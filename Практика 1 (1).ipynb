### 5. Нормализовать собственными функциями:

* столбeц Daily_Usage_Hours порядковым кодированием;

* вывести медиану столбца Impact_on_Grades;

* вывести перцентили столбца Awareness_Level.

```python
def encode_hours(hours):
    if hours < 2:
        return 0 
    elif hours <= 5:
        return 1 
    else:
        return 2  

df['Daily_Usage_Hours_Ordinal'] = df['Daily_Usage_Hours'].apply(encode_hours)



median_val = df['Impact_on_Grades'].median()

p25 = np.percentile(df['Awareness_Level'], 25)
p50 = np.percentile(df['Awareness_Level'], 50)
p75 = np.percentile(df['Awareness_Level'], 75)


print(f"Медиана Impact_on_Grades: {median_val}")
print(f"Перцентили Awareness_Level: 25% = {p25}, 50% = {p50}, 75% = {p75}")
print("\nТаблица после кодирования (первые 5 строк):")
print(df[['Daily_Usage_Hours', 'Daily_Usage_Hours_Ordinal']].head())
```

### 6. Произвести сокращение числа признаков с низкой дисперсией собственной функцией.

```python
import pandas as pd

def filter_low_variance(df, threshold=0.1):
  
    numeric_df = df.select_dtypes(include=['number'])
    
    variances = numeric_df.var()
    
    selected_columns = variances[variances > threshold].index.tolist()
    
    return df[selected_columns]



reduced_df = filter_low_variance(df, threshold=0.1)


print("Таблица признаков с высокой дисперсией:")
print(reduced_df.head())

```
