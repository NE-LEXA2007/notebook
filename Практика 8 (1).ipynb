### 1. Дана функция потерь $ J(w) = \frac{1}{N} \sum{}(y_i - w^tx_i)^2 $, данные: $ X , y $, начальные параметры: $ w $, скорость обучения $ \eta $. Вычислите с помощью собственных функций: 
- Градиент на первом шаге;
- Обновление весов.

```python
import numpy as np

def get_gradient(X, y, w):

    N = X.shape[0]
  
    predictions = X @ w 

    error = predictions - y

    gradient = (2/N) * (X.T @ error)
    return gradient

def update_weights(w, gradient, eta):

    return w - eta * gradient


X = np.array([[1, 2], [1, 3], [1, 4]]) 
y = np.array([5, 7, 9])
w = np.zeros(X.shape[1]) 
eta = 0.01               


grad = get_gradient(X, y, w)
print(f"Градиент на первом шаге: {grad}")


w_new = update_weights(w, grad, eta)
print(f"Обновленные веса: {w_new}")
```

### 2. Для квадратичной функции $ f(w) $ выполните с помощью собственных функций:

- Градиентный спуск со скоростью обучения $ \eta $:
- Метод Ньютона.

```python
import numpy as np

def gradient_descent_step(X, y, w, eta):
  
    N = len(y)
    grad = (2/N) * X.T @ (X @ w - y)
    w_new = w - eta * grad
    return w_new

def newton_method_step(X, y, w):
   
    N = len(y)
 
    grad = (2/N) * X.T @ (X @ w - y)
    
    hessian = (2/N) * (X.T @ X)
    
   
    w_new = w - np.linalg.solve(hessian, grad)
    return w_new


X = np.array([[1, 2], [1, 3], [1, 4]])
y = np.array([5, 7, 9])
w_start = np.array([0.1, 0.1])

w_newton = newton_method_step(X, y, w_start)
print(f"Веса (Ньютон): {w_newton}")


w_gd = gradient_descent_step(X, y, w_start, eta=0.01)
print(f"Веса (Градиентный спуск, 1 шаг): {w_gd}")
```
