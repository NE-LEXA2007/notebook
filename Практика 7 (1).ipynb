### 1. Для $ n $ классов создайте все возможные пары классификаторов ($ \frac{n(n - 1)}{2} $). Обучите несколько моделей для каждой пары с помощью собственных функций.

```python
import numpy as np
from itertools import combinations

def train_binary_model(X, y, eta=0.01, iterations=100):
    w = np.zeros(X.shape[1])
    N = len(y)
    for _ in range(iterations):
        grad = (2/N) * X.T @ (X @ w - y)
        w = w - eta * grad
    return w

def train_ovo_models(X, y, n_classes):
  

    pairs = list(combinations(range(n_classes), 2))
    models = {}

    for (class_a, class_b) in pairs:

        mask = (y == class_a) | (y == class_b)
        X_pair = X[mask]
        y_pair = y[mask]

    
        y_binary = np.where(y_pair == class_a, 0, 1)

      
        print(f"Обучение пары: Класс {class_a} vs Класс {class_b}")
        model_weights = train_binary_model(X_pair, y_binary)
        
        models[(class_a, class_b)] = model_weights
        
    return models


X_mock = np.random.rand(100, 4)
y_mock = np.random.randint(0, 3, 100)

trained_pairs = train_ovo_models(X_mock, y_mock, n_classes=3)
print(f"\nВсего обучено моделей: {len(trained_pairs)}")
```
